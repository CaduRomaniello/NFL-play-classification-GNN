```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carregando os dados do CSV
df = pd.read_csv('../mean_degree.csv')
df = df.iloc[:-1]

# Criando a figura e os eixos
fig, ax1 = plt.subplots(figsize=(12, 8))

# Configurando o primeiro eixo y (mdia)
color1 = 'tab:blue'
ax1.set_xlabel('Nmero de vizinhos mais prximos (n)', fontsize=14)
ax1.set_ylabel('Mdia da Norma de Frobenius', color=color1, fontsize=14)
line1 = ax1.plot(df['n'], df['mean'], marker='o', linestyle='-', 
                 linewidth=2, markersize=8, color=color1, label='Mdia')
ax1.tick_params(axis='y',labelsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)

# Criando o segundo eixo y (desvio padro)
ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Desvio Padro da Norma de Frobenius', color=color2, fontsize=14)
line2 = ax2.plot(df['n'], df['std'], marker='s', linestyle='-', 
                 linewidth=2, markersize=8, color=color2, label='Desvio Padro')
ax2.tick_params(axis='y', labelsize=12)

# Adicionando ttulo
plt.title('Norma de Frobenius vs. Nmero de Vizinhos Mais Prximos', fontsize=16, pad=20)

# Adicionando legenda combinando os dois grficos
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left', fontsize=12)

# # Adicionando anotaes para destacar os valores mximos
# max_mean_index = df['mean_frobenius_norm'].idxmax()
# max_mean_n = df.loc[max_mean_index, 'n']
# max_mean_value = df.loc[max_mean_index, 'mean_frobenius_norm']

# # Anotao para o valor mximo da mdia
# ax1.annotate(f'Mximo: {max_mean_value:.2f}',
#              xy=(max_mean_n, max_mean_value),
#              xytext=(max_mean_n-3, max_mean_value+10),
#              arrowprops=dict(facecolor='black', shrink=0.05, width=1.5),
#              fontsize=12)

# Ajustando o layout para evitar sobreposio
fig.tight_layout()

# Salvando a figura
plt.savefig('../results/mean_degree/frobenius_norm_plot_mean_degree.png', dpi=300, bbox_inches='tight')

# Exibindo o grfico
plt.show()
```
