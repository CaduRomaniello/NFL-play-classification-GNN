```python
import os
import pandas as pd
```

```python
# setting path to data
cur_path = os.path.os.getcwd()
print(cur_path)
data_path = os.path.abspath(os.path.join(cur_path, '../nfl_data/2025/'))
print(data_path)
```

```python
# reading data
games = pd.read_csv(os.path.join(data_path, 'games.csv'))
plays = pd.read_csv(os.path.join(data_path, 'plays.csv'))

games.sort_values(['week'], ascending=[True], inplace=True)
```

```python
# getting first game id
first_game_id = games.iloc[0]['gameId'] # 2022090800
plays['passResult'].unique()
```

```python
games_info = games[['gameId', 'week']]
plays = pd.merge(plays, games_info, on='gameId')
plays.sort_values(['week'], ascending=[True], inplace=True)
plays
```

```python
# getting first game plays
first_game_plays = plays[plays['gameId'] == first_game_id]
```

```python
passes = 0
rushes = 0
both = 0
any = 0
for index, row in plays.iterrows():
    if not pd.isna(row['passLocationType']) and not pd.isna(row['rushLocationType']):
        both += 1
    elif not pd.isna(row['passLocationType']):
        if not pd.isna(row['qbSpike']) and row['qbSpike']:
            raise ValueError('qbSpike is not null and its a pass')
        if not pd.isna(row['qbKneel']) and row['qbKneel']:
            raise ValueError('qbKneel is not null and its a pass')
        if not pd.isna(row['qbSneak']) and row['qbSneak']:
            raise ValueError('qbSneak is not null and its a pass')
        
        passes += 1
    elif not pd.isna(row['rushLocationType']):
        if not pd.isna(row['qbSpike']) and row['qbSpike']:
            raise ValueError('qbSpike is not null and its a rush')
        if not pd.isna(row['qbKneel']) and row['qbKneel'] and row['rushLocationType'] != 'UNKNOWN':
            print(row['qbKneel'], row['playDescription'])
            print(row['rushLocationType'])
            raise ValueError('qbKneel is not null and its a rush')
        if not pd.isna(row['qbSneak']) and row['qbSneak']:
            print(row['qbSneak'], row['playDescription'])
            print(row['passResult'])
            print(row['rushLocationType'])
            raise ValueError('qbSneak is not null and its a rush')
        rushes += 1
    else:
        # if not pd.isna(row['qbSpike']) and row['qbSpike']:
        #     print(f'qbSpike {row['qbSpike']}: ', row['playDescription'])
        # if not pd.isna(row['qbKneel']) and row['qbKneel']:
        #     print(f'qbKneel {row['qbKneel']}: ', row['playDescription'])
        # if not pd.isna(row['qbSneak']) and row['qbSneak']:
        #     print(f'qbSneak {row['qbSneak']}: ', row['playDescription'])
        
        # print()
            
        any += 1
        
print()
print('passes:', passes)
print('rushes:', rushes)
print('both:', both)
print('any:', any)
    
    
    
    
# if pd.isna(play['passLocationType'].values[0]) and pd.isna(play['rushLocationType'].values[0]):
#     raise ValueError('play does not have passLocationType or rushLocationType')
# elif not pd.isna(play['passLocationType'].values[0]) and not pd.isna(play['rushLocationType'].values[0]):
#     raise ValueError('play has both passLocationType and rushLocationType')
# else:
#     playType = 'pass' if pd.isna(play['rushLocationType']).empty else 'rush'
```

```python
ppw = {
    'week_1': {
        'passes': 0,
        'rushes': 0,
    },
    'week_2': {
        'passes': 0,
        'rushes': 0,
    },
    'week_3': {
        'passes': 0,
        'rushes': 0,
    },
    'week_4': {
        'passes': 0,
        'rushes': 0,
    },
    'week_5': {
        'passes': 0,
        'rushes': 0,
    },
    'week_6': {
        'passes': 0,
        'rushes': 0,
    },
    'week_7': {
        'passes': 0,
        'rushes': 0,
    },
    'week_8': {
        'passes': 0,
        'rushes': 0,
    },
    'week_9': {
        'passes': 0,
        'rushes': 0,
    }
}

qbSpike = 0
qbKneel = 0
qbSneak = 0
passes = 0
rushes = 0
scrambles = 0

for index, row in plays.iterrows():
    if not pd.isna(row['qbSpike']) and row['qbSpike']:
        qbSpike += 1
    elif not pd.isna(row['qbKneel']) and row['qbKneel']:
        qbKneel += 1
    elif not pd.isna(row['qbSneak']) and row['qbSneak']:
        qbSneak += 1
    elif row['passResult'] == 'R':
        scrambles += 1
    elif not pd.isna(row['rushLocationType']):
        rushes += 1
        ppw[f'week_{row["week"]}']['rushes'] += 1
    elif not pd.isna(row['passLocationType']):
        passes += 1
        ppw[f'week_{row["week"]}']['passes'] += 1
    elif not pd.isna(row['passResult']):
        print("PASS WITHOUT INFO")
        print(row['playDescription'], ' - ', row['passResult'])
        print(row['qbSpike'], row['qbKneel'], row['qbSneak'], row['passLocationType'], row['rushLocationType'])
        print()
        passes += 1
        ppw[f'week_{row["week"]}']['passes'] += 1
    else:
        print(row['playDescription'], ' - ', row['passResult'])
        print(row['qbSpike'], row['qbKneel'], row['qbSneak'], row['passLocationType'], row['rushLocationType'])
        print()
        
print('qbSpike:', qbSpike)
print('qbKneel:', qbKneel)
print('qbSneak:', qbSneak)
print('passes:', passes)
print('rushes:', rushes)
print('scrambles:', scrambles)

# qbSpike: 23       qbSpike: 23
# qbKneel: 165      qbKneel: 165
# qbSneak: 120      qbSneak: 120
# passes: 9313      passes: 9313
# rushes: 6503      rushes: 6103
# scrambles: 400    scrambles: 400
```
